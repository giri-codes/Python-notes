# Python Booleans:

Booleans means only two possible values.

True

False


```python
# Boolean as a Data Type:
is_raining = True
is_logged_in = False

print(type(is_raining))
```

```python
# Comparison Operators:
print(10>5)
print(4==6)
print(7!=7)
```

```python
# Booleans from Variables:
age = 18
print(age > 18)
```

```python
# Boolean with if Condition:

is_logged_in = True

if is_logged_in:
    print("Welcome user!")
```

```python
# Logical Operators (and, or, not):
# and (Both must be True):

age = 19
has_id = True

print(age >= 18 and has_id)
```

```python
# or (Any one True):

has_cash = False
has_upi = True

print(has_cash or has_upi)
```

```python
# not (Reverse result):
is_active = True
print(not is_active)
```

```python
# bool() Function:
print(bool(1))
print(bool(0))
print(bool("Python"))
print(bool(""))
```

## Practice Examples:

```python
# Boolean from User Input:

answer = input("Do you like Python? (yes/no): ")
print(answer == "yes")

# Comparison converts user input into Boolean.
```

```python
# Equality vs Inequality:

pin = 1234
entered_pin = 1111

print(pin != entered_pin)

# Used to detect wrong input.
```

```python
# Boolean Stored from Calculation:
balance = 500
withdraw = 300

can_withdraw = balance >= withdraw
print(can_withdraw)

# Boolean result stored for later use.
```

```python
# Multiple Conditions with and:

marks = 72
fees_paid = True

print(marks >= 60 and fees_paid)

# Used in college eligibility systems.
```

```python
# Boolean with not and Truthy/Falsy"
otp = ""
print(not otp)

# Empty OTP means not entered.
```

## Real-Life Examples:

```python
# Example 1: ATM Cash Availability:

atm_cash = 10000
requested_cash = 12000

if atm_cash >= requested_cash:
    print("Dispensing Cash")

else:
    print("Insufficient Funds")
```

```python
# Example 2: Online Exam Access:

is_registered = True
exam_time = True

if is_registered and exam_time:
    print("Exam started")
else:
    print("Access denied")
```

```python
# Example 3: Mobile Battery Warning:
battery = 15

print(battery <= 20)

# Phone shows low battery alert.
```

```python
# Example 4: Food Delivery Availability:
restaurant_open = False

if not restaurant_open:
    print("Restaurant closed")
```

```python
# Example 5: Train Ticket Confirmation:
seats_available = 0

print(seats_available > 0)
```
