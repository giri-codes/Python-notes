# Comparison Operators:
“Comparison operators compare operands and always return a Boolean value.”

### Syntax:
result = value1 <`operator`> value2

print(result)


## Python Comparison Operators

| Operator | Name                        | Purpose / Meaning                     |
|--------:|-----------------------------|--------------------------------------|
| `==`    | Equal to                    | Checks if two values are equal        |
| `!=`    | Not equal to                | Checks if two values are not equal    |
| `>`     | Greater than                | Checks if left value is greater       |
| `<`     | Less than                   | Checks if left value is smaller       |
| `>=`    | Greater than or equal to     | Checks if left ≥ right                |
| `<=`    | Less than or equal to        | Checks if left ≤ right                |


```python
# Equal to:
a = 10
b = 10
print(a == b)
```

```python
# Not equal to:
x = 5
y = 8
print(x != y)
```

```python
# Greater than:
age = 21
limit = 18
print(age > limit)
```

```python
# Less than:
marks = 45
pass_marks = 50
print(marks < pass_marks)
```

```python
# Greater than or equal to:
score = 75
cutoff = 75
print(score >= cutoff)
```

```python
# Less than or equal to:
balance = 1000
withdraw = 1000
print(withdraw <= balance)
```

```python
# Expression comparison:
a = 5
b = 2
print((a + b) > (a * b))
```

```python
# String comparison: 
name1 = "apple"
name2 = "banana"
print(name1 < name2)
```

```python
# Variable inside variable:
x = 10
y = x + 5
print(y == 15)
```

```python
# Nested logic result:
a = 20
b = 15
c = 10
print(a > b > c)
```

## Real life examples:

```python
# Voting eligibility:
age = 19
print(age >= 18)
```

```python
# Fuel warning:
fuel = 5
min_fuel = 10
print(fuel < min_fuel)
```

## Real-World Usage of Comparison Operators

1. **User Authentication**
   - Compare entered password with stored password (`==`, `!=`)

2. **Eligibility Checking**
   - Age, marks, experience validation (`>=`, `<`)

3. **Banking & Finance Systems**
   - Balance vs withdrawal amount (`>=`, `<=`)

4. **E-commerce Applications**
   - Stock availability and price comparison (`>`, `==`)

5. **Data Validation & Filtering**
   - Validate ranges in forms, reports, datasets (`<`, `>`, `<=`, `>=`)


## Interview Questions & Answers – Comparison Operators

1. **What is the purpose of comparison operators in Python?**  
   → To compare two values and return a Boolean result (`True` or `False`).

2. **What is the return type of a comparison operation?**  
   → Boolean.

3. **Difference between `=` and `==`?**  
   → `=` assigns a value, `==` compares two values.

4. **Can comparison operators be used with strings?**  
   → Yes, strings are compared lexicographically.

5. **What is the output of `10 == "10"`?**  
   → `False` (different data types).

6. **Which operator checks inequality?**  
   → `!=`

7. **Is chained comparison allowed in Python?**  
   → Yes (example: `10 > 5 > 2`).

8. **Can comparison operators be used inside `if` statements?**  
   → Yes, they are commonly used for decision-making.

9. **Do comparison operators return numeric values?**  
   → No, they return Boolean values only.

10. **Are comparison operators used in loops?**  
    → Yes, especially in `while` loop conditions.
