# ASSIGNMENT OPERATORS:

**Assignment operators are used to assign values to variables.**

| Operator | Meaning                                 |
| -------- | --------------------------------------- |
| `=`      | Assign                                  |
| `+=`     | Add & assign                            |
| `-=`     | Subtract & assign                       |
| `*=`     | Multiply & assign                       |
| `/=`     | Divide & assign                         |
| `//=`    | Floor divide & assign                   |
| `%=`     | Modulus & assign                        |
| `**=`    | Power & assign                          |
| `&=`     | Bitwise AND & assign                    |
| `\|=`    | Bitwise OR & assign                     |
| `^=`     | Bitwise XOR & assign                    |
| `>>=`    | Right shift & assign                    |
| `<<=`    | Left shift & assign                     |
| `:=`     | Assign & return value (Walrus operator) |


```python
# Basic assignment:
a = 5
print(a)
```

```python
# Add & assign:
x = 5
x += 3
print(x)
```

```python
# Subtract & assign:
x = 10
x -= 4
print(x)
```

```python
# Multiply & assign:
x = 4
x *= 3
print(x)
```

```python
# Divide & assign:
x = 20
x /= 4
print(x)
```

```python
# Floor divide & assign:
x = 17
x //= 3
print(x)
```

```python
# Modulus & assign:
x = 17
x %= 5
print(x)
```

```python
# Power & assign:
x = 2
x **= 4
print(x)
```

```python
# Bitwise & and assign:
x = 10

x &= 2

print(x)
```

```python
# Bitwise OR and assign:
x = 8

x |= 4

print(x)
```

```python
# Bitwise XOR & assign:
x = 15

x ^= 9

print(x)
```

```python
# Right shift & assign:
x = 42

x >>= 3

print(x)
```

```python
# Left shift & assign:
x = 42

x <<= 3

print(x)
```

```python
# Assignment inside expression:
x = 5
y = x + 10
print(y)
```

```python
# Chained assignment:
a = b = c = 42
print(a, b, c)
```

# Walrus Operator (:=):

**The walrus operator(:=) allows you to assign a value and use it in the same expression.**

Introduced in Python 3.8

```python
# Basic walrus usage:
print(n := 10)
```

```python
# Inside condition:
if (n := 5) > 3:
    print(n)
```

```python
# Avoid double calculation:
if (length := len("I love python")) > 4:
    print(length)
```

```python
# Input with walrus:
if (num := int(input("Enter: "))) > 10:
    print(num)
```

```python
# While loop usage:
while (n := n-1) > 0:
    print(n)
```

```python
# Walrus with list:
if (items := len([1,2,3])) > 2:
    print(items)
```

```python
# Walrus in expression:
result = (x := 5) + 10
print(result)
```

```python
# Function return capture:
def square(n):
    return n*n

if (res := square(4)) > 10:
    print(res)
```

```python
# Walrus inside loop condition:
numbers = [1,2,3,4,5]
while (n := len(numbers)) > 0:
    numbers.pop()
    print(n)
```

```python
# Advanced filtering:
data = ["hi", "hello", "python"]
if (long := [x for x in data if len(x) > 3]):
    print(long)
```
