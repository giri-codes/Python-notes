## DATA TYPES

### Data types represent the type of value stored inside a variable.

## Data Types in Python

- **Text Type:** `str`
- **Numeric Types:** `int`, `float`, `complex`
- **Sequence Types:** `list`, `tuple`, `range`
- **Mapping Type:** `dict`
- **Set Types:** `set`, `frozenset`
- **Boolean Type:** `bool`
- **Binary Types:** `bytes`, `bytearray`, `memoryview`
- **None Type:** `NoneType`


## Syntax:

```python
# integer
x = 10

# float
y = 3.14

# string
name = "Giri"

# list
fruits = ["apple", "mango", "banana"]

# tuple
values = (10, 20, 30)

# set
unique_numbers = {1, 2, 3}

# dictionary
student = {"name": "Giri", "age": 25}
```

## To get data type:

### Style 1: Direct assignment:

```python
fruit = "Mango"
print(type(fruit))

price = 100
print(type(price))

weight = 1.5
print(type(weight))

g = 8j
print(type(g))

is_fruit = True
is_not_fruit = False

print(f"is_fruit: {is_fruit}, Type: {type(is_fruit)}")
print(f"is_not_fruit: {is_not_fruit}, Type: {type(is_not_fruit)}")
```

### Style 2: Using constructors:

```python
# Example 1:
a = int("42")
b = float("2.5")
c = str(143)

print(type(a))
print(type(b))
print(type(c))
```

```python
# Example 2:
a = 10
b = 3.5
c = "Giri"
d = True

print(type(a))   # <class 'int'>
print(type(b))   # <class 'float'>
print(type(c))   # <class 'str'>
print(type(d))   # <class 'bool'>
```

```python
# Example 3:
total_animals = int("155") #Now it is in string, so by using int we convert it into integer
animals_weight = float("250.5") #str to float
main_animals = str(10) # int to str

print(type(total_animals))
print(type(animals_weight))
print(type(main_animals))
```

### Style 3 : Grouped/Pythonic assignment:

```python
l, m, n = 8, 4.2 , "Gireesh"
print(l)
print(m)
print(n)
```

## Data type Interview Question & Answers:

**Q1. What are Python’s built-in data types?**

Ans : int, float, str, bool, list, tuple, set, dict

**Q2. What is the difference between list and tuple?**

Ans : List is mutable (can change).
      Tuple is immutable (cannot change after creation).

**Q3. What is mutability in Python?**

Ans : Mutability means a value can be changed after creation.
      Mutable types → list, dict, set
      Immutable types → int, float, str, tuple

**Q4. Why do we use a set?**

Ans : Set automatically removes duplicate values and provides fast membership checks.

**Q5. What is a dictionary in Python?**

Ans : A dictionary stores data as key–value pairs.
      Example: {"name": "Giri", "age": 25}


**Q6. Can a list contain multiple data types?**

Ans : Yes. A list can store different types together.
      Example: [10, "Giri", 3.5, True]

**Q7. What happens if you try to modify a tuple?**

Ans : Python throws an error because tuple is immutable.

**Q8. Which data type is best for unique values?**

Ans : Set — because it stores only unique elements.

**Q9. Which data type is best for key–value storage?**

Ans : Dictionary — stores data in key: value format.

**Q10. How do you check the type of a variable?**

Ans : type(x)


## SCENARIO-BASED QUESTIONS WITH ANSWERS:
