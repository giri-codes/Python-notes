# F strings:

```python
# Create an f - string:

name = "Giri"
age = 27
job = "Developer"

text = f" My name is {name} i am {age} and my role is {job}"
print(text)
```

```python
# By using modifiers:

price = 42
txt = f"The price is {price:.2f} rupees"
print(txt)
```

```python
# Placeholder can contain math operations:

price = 42
txt = f"The price of 8 products is {42 * 8} rupees"
print(txt)
```

## Practice Examples:

```python
# Example 1: Full name + age after 5 years
first_name = "Giri"
last_name = "B"
current_age = 25

future_age = current_age + 5
print(f"Name: {first_name} {last_name}, Age after 5 years: {future_age}")
```

```python
# Example 2: Average of 3 numbers
a = 45
b = 55
c = 60

average = (a + b + c) / 3
print(f"Average of {a}, {b}, and {c} is {average:.2f}")
```

```python
# Example 3: Temperature conversion
celsius = 30
fahrenheit = (celsius * 9/5) + 32

print(f"{celsius}°C is equal to {fahrenheit}°F")
```

```python
# Example 4: Length comparison
word1 = "developer"
word2 = "python"

len1 = len(word1)
len2 = len(word2)

print(f"'{word1}' length = {len1}, '{word2}' length = {len2}")
```

```python
# Example 5: Conditional text formatting
num = 17

print(f"Number {num} is {'Even' if num % 2 == 0 else 'Odd'}")
```

## REAL-LIFE f-STRING EXAMPLES:

```python
# Example 6: Employee salary summary
name = "Giri"
basic = 30000
bonus = 5000

total = basic + bonus
print(f"Employee: {name}")
print(f"Total Salary: ₹{total}")
```

```python
# Example 7: Fitness progress
start_weight = 60
current_weight = 64

print(f"Weight Progress: +{current_weight - start_weight} kg")
```

```python
# Example 8: Online order summary
item = "Laptop"
qty = 1
price = 42000

print(f"Order: {item}")
print(f"Payable Amount: ₹{qty * price}")
```

```python
# Example 9: Exam report
student = "Giri"
marks = 78

print(f"Student: {student} | Result: {'Pass' if marks >= 35 else 'Fail'}")
```

```python
# Example 10: Date & time display
day = 5
month = 1
year = 2026
hour = 6
minute = 5

print(f"{day:02}-{month:02}-{year} {hour:02}:{minute:02}")
```
