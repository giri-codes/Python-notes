# Modify Strings :

```python
# Upper Case:
a = "Hi Gireesh"
print(a.upper()) #Here upper() method returns the string in upper case.
```

```python
# Lower Case:
b = "Hi Gireesh"
print(b.lower()) #Here lower() method returns the string in lower case.
```

```python
# Remove whitespace:
c = "  Hello, how are you Gireesh!?    "
print(c.strip()) #Here strip() method removes any whitespace from the beginning or the end.
```

```python
# Replace string:
d = "I am Fine!"
print(d.replace("F", "L")) #Here replace() method replaces a string with another string.
```

```python
# Split string:
e = "I, am a, developer"
print(e.split(",")) #Here split() method splits the string into substrings.
```

## Examples:

```python
# Practice 1: Convert to UPPERCASE
text = "python developer"
result = text.upper()
print(result)
```

```python
# Practice 2: Convert to lowercase
name = "GIRI"
print(name.lower())
```

```python
# Practice 3: Replace a word
sentence = "I am a python learner"
new_sentence = sentence.replace("learner", "developer")
print(new_sentence)
```

```python
# Practice 4: Remove extra spaces (strip)
msg = "   hello everyone   "
print(msg.strip())
```

```python
# Practice 5: Capitalize first letter
line = "python is easy"
print(line.capitalize())
```

```python
# Practice 6: Title case (real life use)
course = "python full stack developer"
print(course.title())
```

```python
# Practice 7: Replace character
phone = "123-456-789"
print(phone.replace("-", ""))
```

```python
# Practice 8: Count characters after modification
text = "Hello World"
modified = text.lower()
print(len(modified))
```

```python
# Practice 9: Remove word
msg = "I love you coding"
print(msg.replace("you ", ""))
```

```python
# Practice 10: Chain multiple modifications
text = "   i am a PYTHON Developer   "
result = text.strip().lower().replace("python", "java")
print(result)
```
